### Setup

```python
import globus_sdk
import json
SDSC_index = '0e8be9d5-99d7-4641-ae43-f72b40bb8a5c'
```

---

### Search a Single Term

```python
sc = globus_sdk.SearchClient()
advanced_result = sc.search(SDSC_index, 'Title: Interactive Computing', limit=1, advanced=True)
print(advanced_result)
```

---

### Search `n` Terms

```python
sc = globus_sdk.SearchClient()
advanced_result = sc.search(SDSC_index, 'Title: Parallel', limit=10, advanced=True)
print(advanced_result)
```

---

### Search and Save to File

```python
sc = globus_sdk.SearchClient()
q = 'Title: Interactive Computing'
advanced_search = sc.search(SDSC_index, q, limit=10, advanced=True)

out_fn = "output_data/search_query.json"
result_formatted = json.loads(advanced_search.text)

json.dump(result_formatted, open(out_fn, 'w'))
```

```python
with open(out_fn, "r") as read_file:
    event = json.load(read_file)
```
